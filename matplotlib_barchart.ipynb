5. Write a Python programming to display a bar chart of the popularity of programming
Languages. Attach a text label above each bar displaying its popularity (float value).<br>

8. Write a Python programming to display a bar chart of the popularity of programming
Languages. Select the width of each bar and their positions.<br>
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def addlabels(language,popularity):
    for counter in range(len(language)):
        bar_width=0.1
        plt.text(counter,popularity[counter],popularity[counter],ha='center')
        plt.bar(language,popularity,color=["green","red","yellow","pink","cyan","blue"],edgecolor="blue",width=bar_width+0.8)
   
language=["Java","Python","PHP","JavaScript","C#","C++"]
popularity=[22.2,17.6,8.8,8,7.7,6.7]
positions = np.arange(len(language))
plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages")
plt.xticks(positions, language)
addlabels(language,popularity)
plt.show()
```

9. Write a Python programming to display a bar chart of the popularity of programming
Languages. Increase bottom margin.<br>
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
def addlabels(language,popularity):
    for counter in range(len(language)):
        bar_width=0.1
        plt.text(counter,popularity[counter],popularity[counter],ha='center')
        plt.bar(language,popularity,color=["green","red","yellow","pink","cyan","blue"],edgecolor="blue",width=bar_width+0.8)
   
language=["Java","Python","PHP","JavaScript","C#","C++"]
popularity=[22.2,17.6,8.8,8,7.7,6.7]
positions = np.arange(len(language))
plt.xlabel("Programming Languages")
plt.ylabel("Popularity (%)")
plt.title("Popularity of Programming Languages")
plt.xticks(positions, language)
addlabels(language,popularity)
plt.subplots_adjust(bottom=0.5)
plt.show()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.<br>
Sample Data:<br>
Means (men) = (22, 30, 35, 35, 26)<br>
Means (women) = (25, 32, 30, 35, 29)

```python
groups = ['A', 'B', 'C', 'D', 'E']
men=np.array([22,30,35,35,26])
women=np.array([25,32,30,35,29])
x = np.arange(len(groups))
width = 0.4
plt.bar(x - width/2,men, width, label='Men', color='blue')
plt.bar(x + width/2,women, width, label='Women', color='pink')
plt.xlabel("Men Vs Women")
plt.ylabel("Mean Value")
plt.show()
```

12. Write a Python program to create bar plots with error bars on the same figure. Sample Date<br>
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824<br>
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
labels = ['A', 'B', 'C', 'D']
fig, ax = plt.subplots()
ax.bar(labels, mean_velocity, yerr=std_dev_velocity, capsize=5, color='skyblue', edgecolor='black')
ax.set_ylabel('Mean Velocity')
ax.set_title('Bar Plot with Error Bars')
plt.show()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text<br>
label above each bar displaying men means (integer value).<br>
Sample Data<br>
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824<br>
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
std_dev_velocity = [0.3314, 0.2278, 0.2836, 0.2645]
labels = ['A', 'B', 'C', 'D']
bars = plt.bar(labels, mean_velocity, yerr=std_dev_velocity, capsize=5, color='skyblue', edgecolor='black')
for bar, mean in zip(bars, mean_velocity):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), str(round(mean)), ha='center', va='bottom')

plt.ylabel('Mean Velocity')
plt.title('Bar Plot with Error Bars')
plt.show()
```
