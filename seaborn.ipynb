<left><h3>Importing the Requirements</h3></left>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

<left><h3>Seaborn Barplot</h3></left>

1. Write a program to draw bar plot of sex against survived for a dataset given in the url<br>
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
df=pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")
sns.barplot(x="sex",y="survived",data=df)
plt.title("Survival Rate by Sex on Titanic")
plt.show()
```

<left><h3>Plotly Scatterplot</h3></left>

1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
x_values=np.random.rand(1000)
y_values=np.random.rand(1000)
df=pd.DataFrame({"x":x_values,"y":y_values})
fig=px.scatter(df,x="x",y="y")
fig.show()
```

2. Write a program to draw line and scatter plots for random 100 x and y coordinates

```python
x_values=np.random.rand(100)
y_values=np.random.rand(100)
df=pd.DataFrame({"x":x_values,"y":y_values})
fig=px.scatter(df,x="x",y="y")
fig.show()
fig=px.line(df,x="x",y="y")
fig.show()
```

3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it.

```python
x_values=np.random.rand(500)
y_values=np.random.rand(500)
df=pd.DataFrame({"x":x_values,"y":y_values})
fig=px.scatter(df,x="x",y="y",title="Styled Scatter Plot",template="plotly_dark")
fig.update_traces(marker=dict(color="red"))
fig.update_layout(plot_bgcolor="cyan",paper_bgcolor="blue")
fig.show()
```

4. Write a program to draw a scatter plot for a given dataset and show datalabels on hover<br>
https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv

```python
df=pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv")
fig=px.scatter(df,y=df["State"],x=df["Population"])
fig.show()
```
